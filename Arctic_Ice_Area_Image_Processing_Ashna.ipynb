Displaying color image and it's properties

```
from google.colab import files
from PIL import Image  
from google.colab.patches import cv2_imshow     #for image display
image= cv2.imread('SSMI_sea_ice_minimum_no_dates.2020.tif')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)                                                               
plt.show()
#find out height and width of original image
print("Image Properties")
print("- Number of Pixels: " + str(image.size))
print("- Shape/Dimensions: " + str(image.shape))
```

Converting colored image into Grayscale

```
# convert to grayscale 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.show()
```

Calculating Arctic Ice Area

```
#getting earth pixels by removing background black pixels
no_black_pix=np.sum(gray==0)
print('- Number of black pixels:',no_black_pix)

number_of_white_pix1 = np.sum(gray >= 210) 
print('- Number of white pixels1:', number_of_white_pix1) 

number_of_gray_pix=np.sum(gray>=128)
print("- Number of gray pixels:",number_of_gray_pix)

#calculation of ice area 
area1=((number_of_white_pix1)/(number_of_gray_pix))*100
print("- area1:",area1)

```
